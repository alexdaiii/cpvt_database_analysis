```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data
x1 = np.random.normal(size=100)
y1 = np.random.normal(size=100)
x2 = np.random.normal(size=100)
y2 = np.random.normal(size=100)

# Long x-tick labels for the second plot
xticks2 = ["LongLabel1", "LongLabel2", "LongLabel3", "LongLabel4"]
xtick_positions = np.linspace(-3, 3, len(xticks2))

# Create a figure with GridSpec for precise layout
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(12, 6))
gs = GridSpec(1, 2, wspace=0.2)

# First plot
ax1 = fig.add_subplot(gs[0, 0])
sns.scatterplot(x=x1, y=y1, ax=ax1)
ax1.set_xlim(-3, 3)
ax1.set_ylim(-3, 3)
ax1.set_aspect('equal')
ax1.set_xticklabels([])  # Remove x-tick labels

# Second plot
ax2 = fig.add_subplot(gs[0, 1])
sns.scatterplot(x=x2, y=y2, ax=ax2)
ax2.set_xlim(-3, 3)
ax2.set_ylim(-3, 3)
ax2.set_aspect('equal')
ax2.set_xticks(xtick_positions)
ax2.set_xticklabels(xticks2, rotation=90, ha="right")

# Adjust layout to align the plotting areas
# plt.tight_layout()
plt.subplots_adjust(hspace=0.1, wspace=0.3)  # Add horizontal/vertical space

plt.show()
```
