```python
import pandas as pd
from analysis.database import get_engine

with get_engine() as engine:
    df = pd.read_sql(
        """
        SELECT structure_id, name AS domain, symbol, parent_id, residue_span
        FROM structure
        """,
        engine
    )
    df["symbol"] = df["symbol"].apply(lambda x: x.strip() if pd.notna(x) else x)

df.to_excel(
    "RYR2_structure.xlsx",
    index=False
)

df.head()
```
