## Calstabin Binding Region

In discussion there is mention of calstabin region which in in these regions:

```
{
[630, 650], 
[680, 690], 
[730, 750], 
[1310,1320],
[1535, 1575],
[1630, 1640],
[1660, 1680],
[1760, 1785]
}
```

````python
import pandas as pd

from analysis.database import get_engine

# just do it in SQL bc it would be terrible to do in regular pandas
with get_engine() as engine:
    cal_binding_variants = pd.read_sql("""
SELECT s.p_hgvs_string, iv.individual_id, ic.age_of_onset, vv.structure_domains
FROM sequence_variant s
JOIN variant v
ON s.sequence_variant_id = v.sequence_variant_id
JOIN individual_variant iv
ON v.variant_id = iv.variant_id
JOIN cpvt_patients_v cp
ON cp.individual_id = iv.individual_id
JOIN individual_condition ic
ON iv.individual_id = ic.individual_id
JOIN variant_view_mv vv
ON v.variant_id = vv.variant_id
WHERE s.p_pos_interval && '{
[630, 650], 
[680, 690], 
[730, 750], 
[1310,1320],
[1535, 1575],
[1630, 1640],
[1660, 1680],
[1760, 1785]
}'::int4multirange
-- NOTE WITHOUT THIS CONDITION WE GET 6 PATIENTS
AND ic.age_of_onset IS NOT NULL
;
""",
                                       engine
                                       )
cal_binding_variants
````
