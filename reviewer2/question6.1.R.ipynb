```r
data_dir <- file.path(getwd(), "..", "data")

fisher_flec_org <- read.csv(file.path(data_dir, "flecainide_fisher_before_2013.csv"))
head(fisher_flec_org)
```

```r
requirements <- c("rstatix", "effectsize", "rcompanion")

for (package in requirements) {
  if (!require(package, character.only = TRUE)) {
    install.packages(package)
  }
}
```

```r
library(rstatix)
```

```r
fisher_flec_test_org <- fisher_flec_org[, c("p_posedit_aa1",
                                            "Flecainide_and_beta_blocker",
                                            "Beta_blocker_only")]
rownames(fisher_flec_test_org) <- fisher_flec_test_org$p_posedit_aa1
fisher_flec_test_org <- fisher_flec_test_org[, -1]
fisher_flec_test_org
```

```r
row_wise_fisher_test(xtab = fisher_flec_test_org, p.adjust.method = "fdr")
```

```r
fisher_org_results = pairwise_fisher_test(x = fisher_flec_test_org, p.adjust.method = "fdr")
# convert to df
fisher_org_results_df <- as.data.frame(fisher_org_results)
# remove scientific notation
options(scipen = 999)
# order by p.value
fisher_org_results_df <- fisher_org_results_df[order(fisher_org_results_df$p.adj),]
# print out only significant results
fisher_sig <- fisher_org_results_df[fisher_org_results_df$p.adj < 0.05,]
fisher_sig
```
