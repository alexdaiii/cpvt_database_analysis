```python


import duckdb

clinvar_duckdb_loc = "/Volumes/T7_Shield/data/2024/cpvt-database/cleanup/data/clinvar.duckdb"

conn = duckdb.connect(clinvar_duckdb_loc)
```

```python
conn.execute("""
SELECT *
FROM clinvar
LIMIT 5
""").fetchdf()
```

```python
# Get the table names
conn.execute("""
SELECT COUNT(*)
FROM clinvar
WHERE Assembly = 'GRCh38'
""").fetchall()
```

```python
# As of 07/2024 there are 2,921,042 variants in the GRCh38 assembly
```

```python
conn.execute("""
SELECT 
    GeneSymbol,
    COUNT(*) AS num_variants
FROM clinvar
WHERE Assembly = 'GRCh38'
GROUP BY GeneSymbol
ORDER BY num_variants DESC
LIMIT 25
""").fetchdf()
```
